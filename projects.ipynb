```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Your Career Aspirations of GenZ.csv")
```

```python
df
```

```python
df.columns
```

```python
class_counts = df['Your Current Country.'].value_counts()

class_counts.plot.pie( startangle=0,shadow=True,figsize=(6,6),ylabel='')
```

```python
country_counts = df['Your Current Country.'].value_counts()
country_counts
```

```python
sns.countplot(x="Your Current Country.",order=country_counts.index,data=df)
plt.xticks(rotation=45)
plt.title("analysis of the country")
plt.xlabel("country")
plt.ylabel("number of responses")
plt.savefig("analysis of the countryt.jpg")

plt.show()
```
